```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = '../data/vgsales.csv'
df = pd.read_csv(path)
df.head()
```

# Global sales by year 

```python
year_s=df.groupby('Year')['Global_Sales'].sum()
print(year_s)
```

# Visualization for Global sales by year

```python
plt.plot(year_s, 'r')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

# Find maximum

```python
int(max(year_s))
```

# Genre Distribution

```python
df['Genre'].value_counts().plot(kind='bar')
plt.title('Genre Distribution')
plt.show()
```

```python
df['Genre'].value_counts()
```

# Global Sales for publisher

```python
p_s=df.groupby('Publisher')['Global_Sales'].sum()
```

# Top 10 global sales for publisher

```python
print(p_s.sort_values(ascending=False).head(10))
```

```python
p_sales_top = p_s.sort_values(ascending=False).head(10)
```

# Top 10 Visualization

```python
p_sales_top.plot(kind='bar')
plt.title('Publisher Sales')
plt.show()
```
